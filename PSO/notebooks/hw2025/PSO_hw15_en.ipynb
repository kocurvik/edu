# Homework Exercise 15 - Turing Patterns

In this homework your task will be to create so-called Turing patterns of images. Turing patterns are created by repeatedly blurring and sharpening an image.

Implement the function `create_turing_pattern`. The parameters are defined in the comments of the function. In each step the function should apply Gaussian blurring with the given parameters and then perform unsharp masking. The steps repeat until a pattern emerges.

For each step the function should perform the following:


1.   $I_{g}^{k+1} = \text{blur}(I^{k}, k, \sigma)$
2.   $I^{k+1} = I_{g}^{k+1} + p (I_{g}^{k+1} - \text{blur}(I_{g}^{k+1}, k, \sigma))$

Where $I^{0}$ is the original image,  $k$ is the gaussian gernel size, $\sigma$ is the standard deviation of the Gaussian kernel and $p$ is the unsharp masking parameter, $\text{blur}$ is the OpenCV `blur` function using the provided parameters. This process should be repeated $n$ times. You may want to clip the images between every iteration.

For the parameters in the test code to make sense, convert the images in your function to floats in the range [0,1]. Running many iterations make take a long time so you can modify the code below and change the resizing while you are working on the algorithm.
