# Morphological operations

In this lab we will work with morphological operations on binary images. In the next lab we will expand this to also include grayscale images.

## Definitions

Let $E \subseteq \mathrm{Z}^d$ be a grid and $A$ is its subset, then this subset is a **binary image**.

**Structural element** $B$ is also a binary image, e.g. $B \subseteq E$.

It can also be **shifted** which we denote as  $B_z = \{b + z | b \in B\}$ for some $z \in E$.


**Erosion** of $A$ with structural element $B$ is defined as $A \ominus B = \{z \in E | B_z \subseteq A \}$

![Erosion](https://upload.wikimedia.org/wikipedia/commons/3/3a/Erosion.png)

The image shows an erosion of a dark blue square by a disk resulting in the light blue square.

**Dilation** of $A$ with structural element $B$ is defined as $A \oplus B = \bigcup_{a \in A} B_a$

![Dilation](https://upload.wikimedia.org/wikipedia/commons/8/8d/Dilation.png)

The image shows an erosion of a dark-blue square by a disk resulting in the light blue shape.

**Closing** of $A$ with structural element $B$ is $(A \oplus B) \ominus B$. Closing usually fills holes in the binary image.

![Closing](https://upload.wikimedia.org/wikipedia/commons/2/2e/Closing.png)

Image shows the closing of the dark-blue shape (union of two squares) by a disk, resulting in the union of the dark-blue shape and the light-blue areas.

**Opening** of $A$ with structural element $B$ is $(A \ominus B) \oplus B$. Opening usually removes smaller objects such as noise from image.

![Opening](https://upload.wikimedia.org/wikipedia/commons/c/c1/Opening.png)

Image shows the opening of the dark-blue square by a disk, resulting in the light-blue square with round corners.

## Properties

**Commutativity** - $A \oplus B = B \oplus A$

**Associativity** - $A \oplus (B \oplus C) = (A \oplus B) \oplus C$

**Shift invariance** - $A \oplus B_z = (A \oplus B)_z$

**Duality** - Erosion and dilation are mutually dual. E.g. erosion of the an is equivalent to inverse of the dilation of its inverse image and vice versa.

## Hit-Miss

Hit-miss transforms the image using two structural elements so that  $HM = I \ominus SE_1 \cap (E / I) \ominus SE_2$. In other words only those pixels remain where $SE_1$ 'fits' a $SE_2$ doesn't.

You can use it using `morpohologyEx`. Note that for hit-miss the kernel has to have both positive and negative values to denote the two structuring elements - see [this OpenCV tutorial](https://docs.opencv.org/4.x/db/d06/tutorial_hitOrMiss.html).
