# Homework 14 - Blurring faces

In this homework we will use OpenCV face detector. The code for it is provided below. Your task is to use it on an image, obtain the bounding boxes of faces and blur them to produce a privacy preserving image.

*Note:* To check out the output format check out the [CascadeClassifier class](https://docs.opencv.org/3.4/d1/de5/classcv_1_1CascadeClassifier.html#aaf8181cb63968136476ec4204ffca498) in OpenCV documentation.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def detect_faces(image_path):
    """
    Detect faces in an image and return bounding boxes.

    Args:
        image_path (str): Path to the input image

    Returns:
        an list of face bounding boxes
    """
    # Read the input image
    img = cv2.imread(image_path)

    # Convert to grayscale (required for Haar Cascade)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Load the pre-trained classifier
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,  # Parameter specifying how much the image size is reduced at each image scale
        minNeighbors=5,   # Parameter specifying how many neighbors each candidate rectangle should have to retain it
        minSize=(30, 30)  # Minimum possible object size
    )

    return faces
```
