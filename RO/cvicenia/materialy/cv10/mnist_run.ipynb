```
import keras
import numpy as np
from keras.datasets import mnist
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/gdrive')
root_path = 'gdrive/My Drive/Colab/'
```

Najprv dáta upravíme do tvaru v akom sme ho dávali sieti.


```
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_test = x_test.astype('float32')
x_test /= 255
```

Načítame model.

```
!ls gdrive/My\ Drive/Colab

model = keras.models.load_model(root_path + 'mnist-fc-trained.h5')

```

A potom s ním môžeme ďalej pracovať. Tu by sme kľudne mohli pokračovať aj s trénovaním.

```
for i in range(x_test.shape[0]):
    x = np.reshape(x_test[i,:,:],(1,784))
    y = model.predict(x)
    best = np.argmax(y[0])
    print(best)
    conf = y[0][best]
    print(conf)
    plt.imshow(x_test[i,:,:])
    plt.show()
```
